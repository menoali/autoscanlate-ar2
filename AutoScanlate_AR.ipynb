```python
# الخلية 3: إعدادات
API_KEY = 'sk-ant-...'  # 👈 ضع مفتاحك هنا
WORKERS = 3
REPACK  = True
print('✅ جاهز')
```

```python
# الخلية 4: رفع الملف
from google.colab import files
print('ارفع: صورة / CBZ / CBR / ZIP')
uploaded = files.upload()
filenames = list(uploaded.keys())
print(f'✅ رُفع: {filenames}')
```

```python
# الخلية 5: ترجمة
import subprocess
for fname in filenames:
    print(f'🚀 {fname}')
    cmd = ['python','autoscanlate.py','--input',fname,'--api_key',API_KEY,'--workers',str(WORKERS)]
    if REPACK: cmd.append('--repack')
    subprocess.run(cmd)
print('✨ انتهى!')
```

```python
# الخلية 6: تحميل النتيجة
import glob, zipfile, os
from google.colab import files
cbz = glob.glob('**/*_ar.cbz', recursive=True)
jpgs = glob.glob('**/*_ar.jpg', recursive=True)
if cbz:
    for f in cbz: print(f'⬇ {f}'); files.download(f)
elif jpgs:
    with zipfile.ZipFile('result.zip','w') as zf:
        for f in sorted(jpgs): zf.write(f,os.path.basename(f))
    files.download('result.zip')
else: print('لم تُعثر على نتائج')
```
